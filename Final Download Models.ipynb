```python
import zipfile
import os

def zip_folder(folder_path, output_path):
    with zipfile.ZipFile(output_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, _, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                arcname = os.path.relpath(file_path, folder_path)
                zipf.write(file_path, arcname=arcname)

# Specify the folder you want to zip and the output zip file name
folder_to_zip = '/content/drive/MyDrive/Colab Notebooks/Voice Cloner/'
output_zip_file = '/content/drive/MyDrive/zips/output.zip'

zip_folder(folder_to_zip, output_zip_file)
print(f"Folder '{folder_to_zip}' has been zipped to '{output_zip_file}'.")
```
